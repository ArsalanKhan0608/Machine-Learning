```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import dendrogram

class HierarchicalClustering:
    def __init__(self, linkage='single'):
        self.linkage = linkage
        self.clusters = None
        self.labels_ = None
        self.children_ = []
        self.distances_ = []

    def fit(self, X):
        self.X = X
        # Initialize each data point as a single cluster
        self.clusters = [[i] for i in range(X.shape[0])]
        distances = pdist(X, metric='euclidean')
        self.distance_matrix = squareform(distances)
        np.fill_diagonal(self.distance_matrix, np.inf)

        self.children_ = []
        self.distances_ = []

        while len(self.clusters) > 1:
            # Find the two closest clusters
            i, j = np.unravel_index(np.argmin(self.distance_matrix, axis=None), self.distance_matrix.shape)
            self._merge_clusters(i, j)

        self.labels_ = self._extract_labels()


    def _merge_clusters(self, i, j):
        new_cluster = self.clusters[i] + self.clusters[j]
        self.children_.append([i, j])
        self.distances_.append(self.distance_matrix[i, j])

        # Update the list of active clusters
        self.clusters.append(new_cluster)
        del self.clusters[max(i, j)], self.clusters[min(i, j)]  # Remove merged clusters

        # Update the distance matrix for the new cluster
        new_distances = np.full(self.distance_matrix.shape[0] - 1, np.inf)  # One less row and column now
        for k, cluster_k in enumerate(self.clusters[:-1]):  # Exclude the newly formed cluster
            if self.linkage == 'single':
                new_distances[k] = min(self.distance_matrix[[min(c1, c2) for c1 in new_cluster for c2 in cluster_k], [max(c1, c2) for c1 in new_cluster for c2 in cluster_k]])
            elif self.linkage == 'complete':
                new_distances[k] = max(self.distance_matrix[[min(c1, c2) for c1 in new_cluster for c2 in cluster_k], [max(c1, c2) for c1 in new_cluster for c2 in cluster_k]])

        # Resize the distance matrix and update with new distances
        self.distance_matrix = np.delete(self.distance_matrix, (j, i), axis=0)  # Remove rows
        self.distance_matrix = np.delete(self.distance_matrix, (j, i), axis=1)  # Remove columns
        self.distance_matrix = np.vstack((self.distance_matrix, new_distances))  # Add new row
        new_distances = np.append(new_distances, np.inf)  # Add infinity for the diagonal
        self.distance_matrix = np.column_stack((self.distance_matrix, new_distances))  # Add new column

        # Update children indices after the merge
        for merge in self.children_:
            for index in range(2):
                if merge[index] > i:
                    merge[index] -= 1
                if merge[index] > j:
                    merge[index] -= 1







    def _extract_labels(self):
        labels = np.zeros(self.X.shape[0], dtype=int)
        for label, cluster in enumerate(self.clusters):
            for index in cluster:
                labels[index] = label
        return labels
def plot_dendrogram(model, **kwargs):
    # Create the linkage matrix
    counts = np.zeros(len(model.children_))
    for i, merge in enumerate(model.children_):
        counts[i] = np.sum([len(model.clusters[child]) for child in merge])
    
    linkage_matrix = np.column_stack([model.children_, model.distances_, counts]).astype(float)

    # Plot the dendrogram
    dendrogram(linkage_matrix, **kwargs)
# Generate sample data
X = np.random.rand(20, 2)  # 20 points in 2D

# Perform hierarchical clustering
model = HierarchicalClustering(linkage='single')
model.fit(X)

# Plot dendrogram
plt.title('Hierarchical Clustering Dendrogram')
plot_dendrogram(model)
plt.xlabel("Index of point")
plt.ylabel("Distance")
plt.show()
```

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from scipy.spatial.distance import pdist
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

class HierarchicalClustering:
    def __init__(self, linkage_method='ward'):
        """
        Initializes the HierarchicalClustering class.
        """
        self.linkage_method = linkage_method

    def fit_predict(self, X):
        """
        Fits the hierarchical clustering model and predicts cluster labels.
        """
        # Calculate the linkage matrix using the chosen method
        Z = linkage(X, method=self.linkage_method)

        # Assign cluster labels
        cluster_labels = fcluster(Z, t=0.7, criterion='distance')
        self.Z = Z  # Store linkage matrix for later use in plotting
        return cluster_labels

    def evaluate(self, X, labels):
        """
        Evaluates the clustering using silhouette score as the cohesion metric.
        """
        score = silhouette_score(X, labels)
        print(f"Silhouette Score: {score}")

    def plot_dendrogram(self):
        """
        Plots a dendrogram to visualize the hierarchical clustering structure.
        """
        plt.figure(figsize=(10, 6))
        dendrogram(self.Z)
        plt.title("Hierarchical Clustering Dendrogram")
        plt.xlabel("Data Points")
        plt.ylabel("Distance")
        plt.show()

# Load the Iris dataset
data = load_iris()
X = data.data

# Initialize and fit the model
model = HierarchicalClustering(linkage_method='average')
labels = model.fit_predict(X)

# Evaluate the model
model.evaluate(X, labels)

# Plot the dendrogram
model.plot_dendrogram()
```
