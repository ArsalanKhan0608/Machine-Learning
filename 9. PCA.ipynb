```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt

def mean_normalize(data):
    return data - np.mean(data, axis=0)

def covariance_matrix(data):
    return np.cov(data.T)

def eigenvectors_and_eigenvalues(cov_mat):
    eigvals, eigvecs = np.linalg.eig(cov_mat)
    idx = eigvals.argsort()[::-1]
    return eigvals[idx], eigvecs[:, idx]

def project_data(data, eigvecs, k):
    return np.dot(data, eigvecs[:, :k])

def reconstruct_data(projected_data, eigvecs, k):
    return np.dot(projected_data, eigvecs[:, :k].T)

def reconstruction_error(original_data, reconstructed_data):
    return np.mean(np.square(original_data - reconstructed_data))

def explained_variance_ratio(eigvals):
    return eigvals / eigvals.sum()

# Load and prepare the Iris dataset
iris = datasets.load_iris()
X = iris.data
X_normalized = mean_normalize(X)

# Calculate PCA components
cov_mat = covariance_matrix(X_normalized)
eigvals, eigvecs = eigenvectors_and_eigenvalues(cov_mat)

# Select top k components
k = 2
X_transformed = project_data(X_normalized, eigvecs, k)
# Evaluation
X_reconstructed = reconstruct_data(X_transformed, eigvecs, k)
rec_error = reconstruction_error(X_normalized, X_reconstructed)
print("Reconstruction Error:", rec_error)

variance_ratios = explained_variance_ratio(eigvals)
print("Explained Variance Ratios:", variance_ratios)

# Visualization
plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=iris.target)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
import numpy as np

def standardize_data(X):
    mean = np.mean(X, axis=0)
    std = np.std(X, axis=0)
    return (X - mean) / std
def compute_covariance_matrix(X):
    return np.cov(X.T)
def eigen_decomposition(cov_matrix):
    eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
    # Sort the eigenvectors by decreasing eigenvalues
    idx = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[idx]
    eigenvectors = eigenvectors[:, idx]
    return eigenvalues, eigenvectors
def select_principal_components(eigenvalues, eigenvectors, k):
    return eigenvectors[:, :k]
def project_data(X, components):
    return np.dot(X, components)
from sklearn.datasets import load_iris

# Load dataset
data = load_iris()
X = data.data

# Standardize the data
X_std = standardize_data(X)

# Compute covariance matrix
cov_matrix = compute_covariance_matrix(X_std)

# Perform eigen decomposition
eigenvalues, eigenvectors = eigen_decomposition(cov_matrix)

# Select top k principal components (let's choose 2 for visualization)
k = 2
components = select_principal_components(eigenvalues, eigenvectors, k)

# Project the data onto the 2D plane of the top 2 components
X_pca = project_data(X_std, components)
def proportion_of_variance_explained(eigenvalues, k):
    total_variance = sum(eigenvalues)
    variance_explained = sum(eigenvalues[:k])
    return variance_explained / total_variance

# Calculate and print the proportion of variance explained
pov = proportion_of_variance_explained(eigenvalues, k)
print(f'Proportion of Variance Explained by the top {k} components: {pov:.2f}')
```
