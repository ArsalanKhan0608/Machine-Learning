```python
import numpy as np

class DBSCAN:
    def __init__(self, eps, minPts):
        self.eps = eps
        self.minPts = minPts
        self.labels = None

    def fit(self, X):
        self.labels = np.zeros(X.shape[0], dtype=int) - 1  # Initialize labels to -1 for noise
        C = 0  # Cluster counter

        for i in range(X.shape[0]):
            if self.labels[i] != -1:  # Already classified
                continue

            neighbors = self.region_query(X, i)

            if len(neighbors) < self.minPts:  # Not a core point
                continue

            C += 1  # Next cluster label
            self.expand_cluster(X, i, neighbors, C)

        return self

    def expand_cluster(self, X, P, neighbors, C):
        self.labels[P] = C
        i = 0
        while i < len(neighbors):
            Pn = neighbors[i]
            if self.labels[Pn] == -1:  # Previously labeled as noise
                self.labels[Pn] = C
            elif self.labels[Pn] == 0:  # Not yet classified
                self.labels[Pn] = C
                PnNeighbors = self.region_query(X, Pn)
                if len(PnNeighbors) >= self.minPts:
                    neighbors = neighbors + PnNeighbors
            i += 1

    def region_query(self, X, P):
        neighbors = []
        for i in range(X.shape[0]):
            if np.linalg.norm(X[i] - X[P]) < self.eps:
                neighbors.append(i)
        return neighbors

    def predict(self, X):
        # In DBSCAN, predict doesn't make sense as in supervised learning models,
        # because DBSCAN doesn't define a predictive model. Instead, we use fit to
        # find the clusters in the training data.
        raise NotImplementedError("DBSCAN does not support prediction.")

def silhouette_score(X, labels):
    from scipy.spatial.distance import pdist, squareform

    # Compute the distance matrix
    distances = squareform(pdist(X))

    A = np.zeros(X.shape[0])
    B = np.zeros_like(A)

    for i in range(X.shape[0]):
        same_cluster = (labels == labels[i])
        other_cluster = (labels != labels[i])

        A[i] = np.mean(distances[i, same_cluster])
        B[i] = np.min([np.mean(distances[i, other_cluster & (labels == unique)]) for unique in np.unique(labels) if unique != labels[i]])

    s = (B - A) / np.maximum(A, B)
    return np.mean(s)

# Example usage with a simple 2D dataset
if __name__ == "__main__":
    from sklearn.datasets import make_moons
    import matplotlib.pyplot as plt

    X, _ = make_moons(n_samples=300, noise=0.1, random_state=42)

    dbscan = DBSCAN(eps=0.2, minPts=5)
    dbscan.fit(X)

    plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels)
    plt.title("DBSCAN Clustering")
    plt.show()

    print("Silhouette Score: ", silhouette_score(X, dbscan.labels))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.spatial.distance import pdist, squareform

class DBSCAN:
    def __init__(self, eps, min_samples, metric='euclidean'):
        self.eps = eps
        self.min_samples = min_samples
        self.metric = metric

    def _calculate_distance(self, p1, p2):
        if self.metric == 'euclidean':
            return np.linalg.norm(p1 - p2)
        elif self.metric == 'manhattan':
            return np.sum(np.abs(p1 - p2))

    def _get_neighborhood(self, point, data):
        neighbors = []
        for i, other_point in enumerate(data):
            if i != point and self._calculate_distance(data[point], other_point) <= self.eps:
                neighbors.append(i)
        return neighbors

    def fit(self, data):
        self.labels_ = np.full(data.shape[0], -1)
        cluster_id = 0

        for point_idx in range(data.shape[0]):
            if self.labels_[point_idx] != -1:
                continue

            neighbors = self._get_neighborhood(point_idx, data)
            if len(neighbors) >= self.min_samples:
                self.labels_[point_idx] = cluster_id

                i = 0
                while i < len(neighbors):
                    neighbor = neighbors[i]
                    if self.labels_[neighbor] == -1:
                        self.labels_[neighbor] = cluster_id

                    new_neighbors = self._get_neighborhood(neighbor, data)
                    if len(new_neighbors) >= self.min_samples:
                        neighbors.extend(new_neighbors)

                    i += 1

                cluster_id += 1

def silhouette_score(X, labels):
    distances = squareform(pdist(X))
    A = np.zeros(X.shape[0])
    B = np.full(X.shape[0], np.inf)

    for i in range(X.shape[0]):
        same_cluster = (labels == labels[i])
        A[i] = np.mean(distances[i, same_cluster])

        for cluster in np.unique(labels[labels != labels[i]]):
            inter_cluster_distance = np.mean(distances[i, labels == cluster])
            B[i] = min(B[i], inter_cluster_distance)

    s = (B - A) / np.maximum(A, B)
    return np.mean(s[s != np.inf])

# Load Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Using only 2 features for visualization

# Fit DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=7, metric='euclidean')
dbscan.fit(X)

# Calculate Silhouette Score
score = silhouette_score(X, dbscan.labels_)
print("Silhouette Score:", score)

# Visualization
plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels_, cmap='viridis', marker='o', edgecolor='k', s=50)
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
